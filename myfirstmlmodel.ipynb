# my first model on ml

# load data

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv')
df
```

# data preparation



### data separation as x and y(a and b)

```python
y = df['logS']
y
```

```python
x = df.drop('logS', axis=1)
x
```

### data splitting

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y,train_size=0.2, random_state=100)
```

```python
x_test
```

```python
x_train
```

# model building

## linear regression

### training the model

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)
```

### applying te model to prediction

```python
y_lr_train_perd = lr.predict(x_train)
y_lr_test_perd = lr.predict(x_test)
```

```python
print (y_lr_train_perd)
```

```python
y_lr_test_perd
```

### evaluate model predection

```python
y_train
```

```python
y_lr_train_perd
```

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_perd)
lr_train_r2 = r2_score(y_train, y_lr_train_perd)

lr_test_mse = mean_squared_error(y_train, y_lr_train_perd)
lr_test_r2 = r2_score(y_test, y_lr_test_perd)
```

```python
print ("LR MSE (TRAIN):", lr_train_mse)
print ("LR R2 (TRAIN):", lr_train_r2)
print ("LR MSE (TEST):", lr_test_mse)
print ("LR R2 (TEST):", lr_test_r2)
```

```python
lr_results = pd.DataFrame(["Linear regrresion", lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
```

```python
lr_results
```

```python
lr_results.columns = ['Method', 'Training MSE', 'Training r2', 'Test MSE', 'Test R2']
```

```python
lr_results
```

## random forest

### training the model

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(x_train, y_train)
```

### applying the model to make a prediction

```python
y_rf_train_perd = rf.predict(x_train)
y_rf_test_perd = rf.predict(x_test)
```

### evaluate model perfromance

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_perd)
rf_train_r2 = r2_score(y_train, y_rf_train_perd)

rf_test_mse = mean_squared_error(y_train, y_rf_train_perd)
rf_test_r2 = r2_score(y_test, y_rf_test_perd)
```

```python
rf_results = pd.DataFrame(["Random forest", rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['Method', 'Training MSE', 'Training r2', 'Test MSE', 'Test R2']
```

```python
rf_results
```

## model comparison

```python
df_models = pd.concat([lr_results, rf_results], axis = 0)
df_models
```
